```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel("D:\Study\Machine Learning\FeatureScaling\dataset1.xlsx")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset["priority"].fillna(dataset["priority"].mean(),inplace=True)
```

```python
dataset["priority"].describe()
```

```python
dataset
```

```python
sns.distplot(dataset["priority"])
plt.show()
```

# Using Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
```

```python
ss.fit(dataset[["priority"]])
```

```python
dataset["priority_ss"]= pd.DataFrame(ss.transform(dataset[["priority"]]),columns=["x"])
```

```python
dataset
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["priority_ss"])
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.distplot(dataset["priority"])
plt.title("Before")
plt.subplot(1,2,2)
sns.distplot(dataset["priority_ss"])
plt.title("After")
plt.show()
```
